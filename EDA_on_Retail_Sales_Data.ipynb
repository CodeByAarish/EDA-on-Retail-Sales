````python

#What is the total sales revenue for the given time period?
total_revenue = sales_data["Total Amount"].sum()
print("The Total Revenue of the Sales is", total_revenue)
````

- As we can see that the Electronics products have the Highest Sales Revenue in the Entire Product Category.
- Beauty Products have the lowest Sales Revenue.

````python
# Can we identify any seasonality or trends in sales throughout the year?
sales_trend = sales_data.groupby('Date')['Total Amount'].sum().sort_values(ascending = False)
sales_trend
````

- The Maximum sales is on 23rd May 2023
- The Minimum sales is on 17th April 2023

````python
# Group the sales data by date and calculate the total sales for each date
sales_trend = sales_data.groupby('Date')['Total Amount'].sum().sort_index()

# Plot the line plot
plt.figure(figsize=(10, 6))
sales_trend.plot(kind='line', color='green')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
````

````python
# Filter sales data for the years 2023 and 2024
sales_data_2023_2024 = sales_data[sales_data['Date'].dt.year.isin([2023, 2024])]

# Group by year and calculate total sales
yearly_sales = sales_data_2023_2024.groupby(sales_data_2023_2024['Date'].dt.year)['Total Amount'].sum()

# Plot the total sales for each year using a bar graph
yearly_sales.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales in 2023 and 2024')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
````

````python
#We need to do the Analysis and Visualize the Monthly Sales

# Group by year and month and calculate total sales
year_monthly_sales = sales_data_2023_2024.groupby([sales_data_2023_2024['Date'].dt.year, sales_data_2023_2024['Date'].dt.month])['Total Amount'].sum()

# Step 4: Plot the total sales for each year and month using a bar graph
year_monthly_sales.unstack().plot(kind='bar', figsize=(12, 6))
plt.title('Monthly Sales in 2023 and 2024')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.legend(title='Year')
plt.grid(axis='y')
plt.show()
````

````python
# What is the average price per unit for each product category?
avg_price_per_unit = sales_data.groupby('Product Category')['Price per Unit'].mean().sort_values(ascending = False)
print("The Average Price per Unit of Different Product Category is:")
avg_price_per_unit
````

````python
avg_price_per_unit.plot(kind = 'bar', figsize = (12,6), color = colors)
colors = sns.color_palette('pastel')
plt.xlabel('Product Category')
plt.ylabel('Average Price per Unit')
plt.title('Average Price per Unit of Different Product Category')
plt.xticks(rotation = 0)
plt.show()
````

````python
#  How does the total amount spent vary with age group?
#First we have to create Age Group Boundaries
age_bins = [0, 20, 30, 40, 50, 60, 70, 100]
#Age Groups
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+']
#We can use Groupby method to find out the required Sales
sales_data['Age Group'] = pd.cut(sales_data['Age'], bins = age_bins, labels = age_labels, right = False)
age_group_wise_sales = sales_data.groupby('Age Group')['Total Amount'].sum()
age_group_wise_sales
````

````python
age_group_wise_sales.plot(kind = 'bar', figsize = (12,6), color = colors)
colors = sns.color_palette('pastel')
plt.xlabel('Age Groups')
plt.ylabel('Sales Amount')
plt.title('Sales as per different Age Groups')
plt.xticks(rotation = 0)
plt.show()
````

- As we can see that the maximun sales which we get is from (51-60) Ages Group Peoples
- The minimun sales is from (0-20) Aged Group Teenagers

````python
# What is the distribution of purchases across different days of the week?
# Extract the day of the week from the 'Date' column
sales_data['Day of Week'] = sales_data['Date'].dt.day_name()

# Group by day of the week and calculate total number of purchases
purchases_by_day = sales_data.groupby('Day of Week').size()
purchases_by_day
````

````python
#Show the above Analysis on the Bar Graph
purchases_by_day.plot(kind = 'bar', figsize = (12,6), color = colors)
colors = sns.color_palette('pastel')
plt.xlabel('Day of Week')
plt.ylabel('Number of Purchases')
plt.title('Distribution of Purchases Across Different Days of the Week')
plt.xticks(rotation = 45)
plt.show()
````

- As we can see that the Number of Purchases is Highest on Tuesday and lowest on Thursday

````python
# Can we identify any correlations between the quantity purchased and the total amount spent?

#Calculate the Pearson correlation coefficient
correlation = sales_data['Quantity'].corr(sales_data['Total Amount'])

#Visualize the relationship using a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(sales_data['Quantity'], sales_data['Total Amount'], color='red')
plt.title('Quantity vs Total Amount')
plt.xlabel('Quantity Purchased')
plt.ylabel('Total Amount Spent')
plt.grid(True)
plt.show()

print("Pearson correlation coefficient:", correlation)
````

- As we can see that the points in the scatter plot form an upward trend from left to right, it suggests a positive correlation. This means that as the quantity purchased increases, the total amount spent also tends to increase.
- As the Cofficient of Correlation is around 0.3737 which is a positive value and shows the strong Positive Relationship between both Variables

````python
#Heatmap for the Data

# Calculate the correlation matrix for the entire dataset
correlation_matrix = sales_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Sales Data')
plt.show()
````
